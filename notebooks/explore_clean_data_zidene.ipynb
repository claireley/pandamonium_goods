```python
##############################################################################
# create the user table
##############################################################################
def user_table(df):
    df1 = df.copy()

    id_lists = df1["user_id"].apply(
                                    lambda x : x.split(",")
                                    )
    name_lists = df1["user_name"].apply(
                                        lambda x : x.split(",")
                                        )
    #Extrat id and name list
    id_list = [str(id).strip() for id_list in id_lists for id in id_list]
    name_list = [str(name).strip() for nam_list in name_lists for name in nam_list]
    user_id_name_tuples = list(
                                zip(id_list, name_list)
                              )

    user_df = pd.DataFrame(user_id_name_tuples)
    #determine the unique user ID
    user_df.insert(
                    0, "id_user",
                    range(1, len(user_df) + 1)
                  )
    user_df.columns = ["id_user", "user_code", "user_name"]
    #user_df["id_name"] = user_df["user_id"] + "_" + user_df["user_name"]
    return user_df

##############################################################################
# creat the view table
##############################################################################
def viewer_table(df):
    """
    input: cleaned df
    create the viewer table
    """
    df1 = df.copy()
    review_code_lists = df1["review_id"].apply(lambda x : str(x).split(","))
    review_title_lists = df1["review_title"].apply(lambda x : str(x).split(","))
    review_cont_lists = df1["review_content"].apply(lambda x : str(x).split(","))
    #Extract individual code, title, review
    review_code_list =[str(code).strip() for lis in review_code_lists for code in lis]
    review_title_list =[str(title).strip() for lis in review_title_lists for title in lis]
    review_cont_list =[str(cont).strip() for lis in  review_cont_lists for cont in lis]

    viewer_df = pd.DataFrame(zip(review_code_list, review_title_list, review_cont_list))
    viewer_df.columns = ["viwer_code", "viwer_title", "viewer_content"]
    viewer_df.insert(0, "id_viewer", range(1, len(review_code_list) + 1))
    return viewer_df
##############################################################################
# creat the product table
##############################################################################
def product_table(df):
    """
    input: cleaned df
    create the viewer table
    """
    df1 = df.copy()
    df1.product_id = df1["product_id"].apply(lambda x : str(x).strip())
    df1.product_name = df1["product_name"].apply(lambda x : str(x).strip())
    df_product= (df1[["product_id", 
                      "product_name",
                      "sub_cat1",
                      "sub_cat_2",
                      "discounted_price",
                      "actual_price",
                      "discount_percentage",
                      "about_product",
                      "rating",
                      "rating_count"
                      ]]
                )
    #check uniqueness of product_id
    #df_product["prod_id_name"] = df_product["product_id"] +"_"+ df_product["product_name"]
    df_product.drop_duplicates(subset = "product_id" , keep = "first", inplace = True, ignore_index = True)
    return df_product


df1 = product_table(df)
df1.prod_id_name.duplicated().sum()

```
